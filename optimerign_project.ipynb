```python
from __future__ import print_function
import os
import csv
import pandas as pd
 
path = './helsinki'
list_ = []
files = os.listdir(path)

for name in files:
    if (name != '.DS_Store'):
        file = os.listdir('./helsinki/' + name)
        for f in file:
            frame = pd.read_csv('./helsinki/' + name + '/' + f, sep=';') 
            list_.append(frame)

df = pd.concat(list_, axis = 0, ignore_index = True)
df
```
